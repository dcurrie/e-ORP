```python
import ipywidgets as widgets
import IPython
from IPython.display import display, HTML

import pandas as pd

## Unit Test Code for e-ORP

# display(HTML(f'<strong style="color:royalblue;"> WARNING! </strong> {str}'))

# input widget for file name for saving dataframe
efname = widgets.Text(
    value='../data/_explore.csv',
    placeholder='../data/_explore.csv',
    description='Test output at:',
    style={'description_width': 'initial'},
    disabled=False)

# purpose output for text, tables, graphs, etc.
out_box = widgets.Output(layout={'border': '1px solid black'})
err_out = widgets.Output(layout={'border': '1px solid black'})


# test framework

q_ok_tests_tot = 0
q_ok_tests_sub = 0
q_ng_tests_tot = 0
q_ng_tests_sub = 0

test_name = ''
subtest_name = ''
report_strs = ''

def begin_test(name):
    global test_name
    global subtest_name
    global q_ok_tests_tot
    global q_ng_tests_tot
    global q_ok_tests_sub
    global q_ng_tests_sub
    global report_strs
    q_ok_tests_tot = 0
    q_ok_tests_sub = 0
    q_ng_tests_tot = 0
    q_ng_tests_sub = 0
    test_name = ''
    subtest_name = ''
    report_strs = ''
    test_name = name

def begin_subtest(name):
    global subtest_name
    global q_ok_tests_sub
    global q_ng_tests_sub
    global report_strs
    q_ok_tests_sub = 0
    q_ng_tests_sub = 0
    report_strs = ''
    subtest_name = name

def ok_():
    global q_ok_tests_tot
    global q_ok_tests_sub
    q_ok_tests_tot += 1
    q_ok_tests_sub += 1
    return True

def ng_(s):
    global q_ng_tests_tot
    global q_ng_tests_sub
    global report_strs
    q_ng_tests_tot += 1
    q_ng_tests_sub += 1
    report_strs += s
    return False
   
def chk_eq(tn, a, b):
    if a == b:
        return ok_()
    else:
        return ng_(f'#{tn} neq: {a} {b}\n')

def chk_ae(tn, a, b): # approximately equal... round to the dollar
    if round(a,3) == round(b,3):
        return ok_()
    else:
        return ng_(f'#{tn} nae: {a} {b}\n')

def chk_re(tn, a, b): # roughly equal... round to ten dollars
    if round(a,2) == round(b,2):
        return ok_()
    else:
        return ng_(f'#{tn} nae: {a} {b}\n')

def chk_le(tn, a, b):
    if a <= b:
        return ok_()
    else:
        return ng_(f'#{tn} nle: {a} {b}\n')

def report_sub():
    with out_box:
        display(HTML(f'<strong style="color:blue;"> {subtest_name} </strong>'))
        if q_ng_tests_sub != 0:
            display(HTML(f'<div style="color:red;"> failed {q_ng_tests_sub} </div>'))
            print(report_strs)
        if q_ok_tests_sub != 0:
            display(HTML(f'<div style="color:green;"> passed {q_ok_tests_sub} </div>'))

def report_tot():
    with out_box:
        display(HTML(f'<strong style="color:magenta;"> {test_name} </strong>'))
        if q_ng_tests_tot != 0:
            display(HTML(f'<div style="color:red;"> failed {q_ng_tests_tot} </div>'))
        if q_ok_tests_tot != 0 or q_ng_tests_tot == 0:
            display(HTML(f'<div style="color:green;"> passed {q_ok_tests_tot} </div>'))

# begin_test("Kahuna")
# begin_subtest("Test 1")
# q_ng_tests_sub = 0
# q_ok_tests_sub = 11
# report_sub()
# report_tot()

# 'e'
# 'j'
# 'afterTax'
# 'e_Roth'
# 'e_Taxd'
# 'j_Roth'
# 'j_Taxd'
# 'e_RMD'
# 'j_RMD'
# 'to_aTax'
# 'from_aTax'
# 'from_eRoth'
# 'from_jRoth'
# 'from_eTaxd'
# 'from_jTaxd'
# 'e_RothConv'
# 'j_RothConv'
# 'SSA_income'
# 'pension_income'
# 'misc_income'
# 'charity'
# 'e_Taxd_in'
# 'j_Taxd_in'
# 'spend_essence'
# 'taxfree_income'
# 'auto_income'
# 'disp_income'
# 'taxable_income'
# 'IRMAA'
# 'dividends'
# 'capgains'
# 'caplosss'
# 'QCD_limit'
# 'QCD'
# 'income_tax'
# 'tax_bracket'
# 'cgains_rate'
# 'MAGI'
# 'ror_bonds'
# 'ror_stock'
# 'dvd_stock'
# 'frac_bonds_a'
# 'frac_stock_a'
# 'frac_bonds_r'
# 'frac_stock_r'
# 'frac_bonds_d'
# 'frac_stock_d'
# 'aTax__cash'
# 'aTax_bonds'
# 'aTax_basis'
# 'aTax_unrlz'
# 'fm_aTax__cash'
# 'fm_aTax_bonds'
# 'fm_aTax_basis'
# 'fm_aTax_unrlz'
# 'to_aTax__cash'
# 'to_aTax_bonds'
# 'to_aTax_basis'
# 'to_aTax_unrlz'
# 'e_RMD_factor'
# 'j_RMD_factor'
# 'tax0'
# 'tax1'
# 'tax2'
# 'tax3'
# 'tax4'
# 'tax5'
# 'tax6'
# 'cgt0'
# 'cgt15'
# 'obbba_pax'
# 'IRMAA-buk0'
# 'IRMAA-buk1'
# 'IRMAA-buk2'
# 'IRMAA-buk3'
# 'IRMAA-buk4'
# 'IRMAA-buk5'
# 'IRMAA-chg0'
# 'IRMAA-chg1'
# 'IRMAA-chg2'
# 'IRMAA-chg3'
# 'IRMAA-chg4'
# 'IRMAA-chg5'
# 'IRMAA-bins'
# 'spend_Î´'
# 'surplus'
# 'net_pretax'
# 'net_postax'
# 'year'

def test(_):
    """test _explore file and display results"""
    err_out.clear_output() # at start of each run
    out_box.clear_output() # at start of each run
    #
    begin_test(f'Checked {efname.value}')
    #
    dd = pd.read_csv(efname.value) # , index_col='year'
    YRS = len(dd['e']) - 1 # number of years of projection from base year 0
    # squirreled away values
    infl = dd['IRMAA-buk0'][0] 
    y_spouse_leaves_plan = dd['IRMAA-buk3'][0]
    filing_status = round(dd['tax_bracket'][0])
    # with err_out:
    #     print(f"y: {y_spouse_leaves_plan} fs: {filing_status}")
    for y in range(1,YRS+1):
        year = dd['year'][y]
        begin_subtest(f'Year {year}')
        
        income = dd['taxable_income'][y]
        capgains = dd['capgains'][y]
        ssa_income = dd['SSA_income'][y]
        MAGI = income + capgains + 0.15 * ssa_income + dd['QCD'][y]
        # check_ae fails with some $2 errors...
        chk_re(101, 0, # income
                         dd['e_RMD'][y]
                       + dd['j_RMD'][y]
                       + dd['SSA_income'][y]
                       + dd['misc_income'][y]
                       + dd['pension_income'][y]
                       + dd['taxfree_income'][y]
                       + dd['dividends'][y]
                       # withdrawls
                       + dd['from_aTax'][y]
                       + dd['from_eTaxd'][y]
                       + dd['from_jTaxd'][y]
                       + dd['from_eRoth'][y]
                       + dd['from_jRoth'][y]
                       + dd['e_RothConv'][y]
                       + dd['j_RothConv'][y]
                       # transfers
                       - dd['to_aTax'][y]
                       - dd['e_RothConv'][y]
                       - dd['j_RothConv'][y]
                       # spending
                       - dd['IRMAA'][y]
                       - dd['income_tax'][y]
                       - dd['disp_income'][y])
        chk_le(102, dd['QCD'][y], dd['QCD_limit'][y])
        report_sub()
        #with out_box:
        #    if True:
        #        print(f"{dd['year'][y]} ok")
        #    else:
        #        print(f"{dd['year'][y]} ti {income} cg {capgains} it {ibtax} ct {cgtax}")
        #        print(f"{dd['year'][y]} tc {tax} eo {dd['income_tax'][y]} fs {filing_status} ng")
    report_tot()


tst_button = widgets.Button(description='Test e-ORP', disabled=False,)
tst_button.on_click(test)

winputs = widgets.VBox([
    efname, tst_button,
    err_out,
    out_box
])

display(winputs)
```
