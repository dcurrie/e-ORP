```python
import ipywidgets as widgets
import IPython
from IPython.display import display, HTML

import pandas as pd

## Unit Test Code for e-ORP

# display(HTML(f'<strong style="color:royalblue;"> WARNING! </strong> {str}'))

# input widget for file name for saving dataframe
efname = widgets.Text(
    value='../data/_explore.csv',
    placeholder='../data/_explore.csv',
    description='Test output at:',
    style={'description_width': 'initial'},
    disabled=False)

# purpose output for text, tables, graphs, etc.
out_box = widgets.Output(layout={'border': '1px solid black'})
err_out = widgets.Output(layout={'border': '1px solid black'})


# test framework

q_ok_tests_tot = 0
q_ok_tests_sub = 0
q_ng_tests_tot = 0
q_ng_tests_sub = 0

test_name = ''
subtest_name = ''
report_strs = ''

def begin_test(name):
    global test_name
    global subtest_name
    global q_ok_tests_tot
    global q_ng_tests_tot
    global q_ok_tests_sub
    global q_ng_tests_sub
    global report_strs
    q_ok_tests_tot = 0
    q_ok_tests_sub = 0
    q_ng_tests_tot = 0
    q_ng_tests_sub = 0
    test_name = ''
    subtest_name = ''
    report_strs = ''
    test_name = name

def begin_subtest(name):
    global subtest_name
    global q_ok_tests_sub
    global q_ng_tests_sub
    global report_strs
    q_ok_tests_sub = 0
    q_ng_tests_sub = 0
    report_strs = ''
    subtest_name = name

def ok_():
    global q_ok_tests_tot
    global q_ok_tests_sub
    q_ok_tests_tot += 1
    q_ok_tests_sub += 1
    return True

def ng_(s):
    global q_ng_tests_tot
    global q_ng_tests_sub
    global report_strs
    q_ng_tests_tot += 1
    q_ng_tests_sub += 1
    report_strs += s
    return False
   
def chk_eq(tn, a, b):
    if a == b:
        return ok_()
    else:
        return ng_(f'#{tn} neq: {a} {b}\n')

def chk_ae(tn, a, b): # approximately equal... round to the dollar
    if round(a,3) == round(b,3):
        return ok_()
    else:
        return ng_(f'#{tn} nae: {a} {b}\n')

def chk_re(tn, a, b): # roughly equal... within two dollars fifty cents; needed due to rounding by e-ORP export code
    if abs(a - b) < 0.0025:
        return ok_()
    else:
        return ng_(f'#{tn} nae: {a} {b}\n')

def chk_le(tn, a, b):
    if a <= b:
        return ok_()
    else:
        return ng_(f'#{tn} nle: {a} {b}\n')

def report_sub():
    with out_box:
        display(HTML(f'<strong style="color:blue;"> {subtest_name} </strong>'))
        if q_ng_tests_sub != 0:
            display(HTML(f'<div style="color:red;"> failed {q_ng_tests_sub} </div>'))
            print(report_strs)
        if q_ok_tests_sub != 0:
            display(HTML(f'<div style="color:green;"> passed {q_ok_tests_sub} </div>'))

def report_tot():
    with out_box:
        display(HTML(f'<strong style="color:magenta;"> {test_name} </strong>'))
        if q_ng_tests_tot != 0:
            display(HTML(f'<div style="color:red;"> failed {q_ng_tests_tot} </div>'))
        if q_ok_tests_tot != 0 or q_ng_tests_tot == 0:
            display(HTML(f'<div style="color:green;"> passed {q_ok_tests_tot} </div>'))

# begin_test("Kahuna")
# begin_subtest("Test 1")
# q_ng_tests_sub = 0
# q_ok_tests_sub = 11
# report_sub()
# report_tot()

# 'e'
# 'j'
# 'afterTax'
# 'e_Roth'
# 'e_Taxd'
# 'j_Roth'
# 'j_Taxd'
# 'e_RMD'
# 'j_RMD'
# 'to_aTax'
# 'from_aTax'
# 'from_eRoth'
# 'from_jRoth'
# 'from_eTaxd'
# 'from_jTaxd'
# 'e_RothConv'
# 'j_RothConv'
# 'SSA_income'
# 'pension_income'
# 'misc_income'
# 'charity'
# 'e_Taxd_in'
# 'j_Taxd_in'
# 'spend_essence'
# 'taxfree_income'
# 'auto_income'
# 'disp_income'
# 'taxable_income'
# 'IRMAA'
# 'dividends'
# 'capgains'
# 'caplosss'
# 'QCD_limit'
# 'QCD'
# 'income_tax'
# 'tax_bracket'
# 'cgains_rate'
# 'MAGI'
# 'ror_bonds'
# 'ror_stock'
# 'dvd_stock'
# 'frac_bonds_a'
# 'frac_stock_a'
# 'frac_bonds_r'
# 'frac_stock_r'
# 'frac_bonds_d'
# 'frac_stock_d'
# 'aTax__cash'
# 'aTax_bonds'
# 'aTax_basis'
# 'aTax_unrlz'
# 'fm_aTax__cash'
# 'fm_aTax_bonds'
# 'fm_aTax_basis'
# 'fm_aTax_unrlz'
# 'to_aTax__cash'
# 'to_aTax_bonds'
# 'to_aTax_basis'
# 'to_aTax_unrlz'
# 'e_RMD_factor'
# 'j_RMD_factor'
# 'tax0'
# 'tax1'
# 'tax2'
# 'tax3'
# 'tax4'
# 'tax5'
# 'tax6'
# 'cgt0'
# 'cgt15'
# 'obbba_pax'
# 'IRMAA-buk0'
# 'IRMAA-buk1'
# 'IRMAA-buk2'
# 'IRMAA-buk3'
# 'IRMAA-buk4'
# 'IRMAA-buk5'
# 'IRMAA-chg0'
# 'IRMAA-chg1'
# 'IRMAA-chg2'
# 'IRMAA-chg3'
# 'IRMAA-chg4'
# 'IRMAA-chg5'
# 'IRMAA-bins'
# 'spend_Î´'
# 'surplus'
# 'net_pretax'
# 'net_postax'
# 'year'

def test(_):
    """test _explore file and display results"""
    err_out.clear_output() # at start of each run
    out_box.clear_output() # at start of each run
    #
    begin_test(f'Checked {efname.value}')
    #
    dd = pd.read_csv(efname.value) # , index_col='year'
    #
    def rori_r(y): # Roth
        return 1.0 + (dd['ror_stock'][y] * dd['frac_stock_r'][y] + dd['ror_bonds'][y] * dd['frac_bonds_r'][y])
    def rori_d(y): # TaxD
        return 1.0 + (dd['ror_stock'][y] * dd['frac_stock_d'][y] + dd['ror_bonds'][y] * dd['frac_bonds_d'][y])
    #
    YRS = len(dd['e']) - 1 # number of years of projection from base year 0
    # squirreled away values
    infl = dd['IRMAA-buk0'][0] 
    y_spouse_leaves_plan = dd['IRMAA-buk3'][0]
    filing_status = round(dd['tax_bracket'][0])
    # with err_out:
    #     print(f"y: {y_spouse_leaves_plan} fs: {filing_status}")
    for y in range(1,YRS+1):
        year = dd['year'][y]
        begin_subtest(f'Year {year}')
        
        income = dd['taxable_income'][y]
        capgains = dd['capgains'][y]
        ssa_income = dd['SSA_income'][y]
        MAGI = income + capgains + 0.15 * ssa_income + dd['QCD'][y]
        # check_ae fails with some $2 errors...
        chk_re(100,    # income
                         dd['e_RMD'][y]
                       + dd['j_RMD'][y]
                       + dd['SSA_income'][y]
                       + dd['misc_income'][y]
                       + dd['pension_income'][y]
                       + dd['taxfree_income'][y]
                       + dd['dividends'][y]
                       # withdrawls
                       + dd['from_aTax'][y]
                       + dd['from_eTaxd'][y]
                       + dd['from_jTaxd'][y]
                       + dd['from_eRoth'][y]
                       + dd['from_jRoth'][y]
                       + dd['e_RothConv'][y]
                       + dd['j_RothConv'][y],
                       # transfers
                         dd['to_aTax'][y]
                       + dd['e_RothConv'][y]
                       + dd['j_RothConv'][y]
                       # spending
                       + dd['IRMAA'][y]
                       + dd['income_tax'][y]
                       + dd['disp_income'][y])
        chk_le(101, dd['QCD'][y], round(dd['charity'][y], 3))
        chk_le(102, dd['QCD'][y], dd['QCD_limit'][y])
        chk_re(103, dd['e_Taxd'][y], 
                       rori_d(y) *
                         (   dd['e_Taxd'][y-1]
                           - dd['e_RMD'][y]
                           - dd['e_RothConv'][y]
                           - dd['from_eTaxd'][y]
                           + dd['e_Taxd_in'][y])) # lump sum pension distribution
        chk_re(104, dd['j_Taxd'][y], 
                       rori_d(y) *
                         (   dd['j_Taxd'][y-1]
                           - dd['j_RMD'][y]
                           - dd['j_RothConv'][y]
                           - dd['from_jTaxd'][y]
                           + dd['j_Taxd_in'][y])) # lump sum pension distribution
        chk_re(105, dd['e_Roth'][y], 
                       rori_r(y) * 
                         (  dd['e_Roth'][y-1] - dd['from_eRoth'][y] + dd['e_RothConv'][y]))
        chk_re(106, dd['j_Roth'][y], 
                       rori_r(y) * 
                         (  dd['j_Roth'][y-1] - dd['from_jRoth'][y] + dd['j_RothConv'][y]))
        chk_re(107, dd['aTax_unrlz'][y], dd['aTax_unrlz'][y-1] + dd['to_aTax_unrlz'][y] - dd['fm_aTax_unrlz'][y])
        chk_re(108, dd['aTax_basis'][y], dd['aTax_basis'][y-1] + dd['to_aTax_basis'][y] - dd['fm_aTax_basis'][y])
        chk_re(109, dd['aTax_bonds'][y], dd['aTax_bonds'][y-1] + dd['to_aTax_bonds'][y] - dd['fm_aTax_bonds'][y])
        chk_re(110, dd['aTax__cash'][y], dd['aTax__cash'][y-1] + dd['to_aTax__cash'][y] - dd['fm_aTax__cash'][y])
        chk_re(111, dd['to_aTax_unrlz'][y], (dd['ror_stock'][y] - dd['dvd_stock'][y]) 
                                               * (dd['aTax_basis'][y-1] + dd['aTax_unrlz'][y-1]))
        chk_re(112, dd['dividends'][y], dd['ror_bonds'][y] * dd['aTax_bonds'][y-1]
                                            + dd['dvd_stock'][y] * (dd['aTax_basis'][y-1] + dd['aTax_unrlz'][y-1]))
        # infl
        # TODO -- not yet handled: pension COLAs, phases of retirement event adjustments
        if y > 1:
            chk_ae(113, dd['QCD_limit'][y], (dd['QCD_limit'][y-1] / (2 if y == y_spouse_leaves_plan else 1) * (1 + infl)))
            chk_ae(114, dd['IRMAA-buk0'][y], (dd['IRMAA-buk0'][y-1] / (2 if y == y_spouse_leaves_plan + 2 else 1) * (1 + infl)))
            chk_ae(115, dd['IRMAA-buk1'][y], (dd['IRMAA-buk1'][y-1] / (2 if y == y_spouse_leaves_plan + 2 else 1) * (1 + infl)))
            chk_ae(116, dd['IRMAA-buk2'][y], (dd['IRMAA-buk2'][y-1] / (2 if y == y_spouse_leaves_plan + 2 else 1) * (1 + infl)))
            chk_ae(117, dd['IRMAA-buk3'][y], (dd['IRMAA-buk3'][y-1] / (2 if y == y_spouse_leaves_plan + 2 else 1) * (1 + infl)))
            # IRMAA levels 4 & 5 behave differently if filing status changes, it's not as simple as half
            chk_ae(118, dd['IRMAA-buk4'][y], (dd['IRMAA-buk4'][y-1] / ((35/30) if y == y_spouse_leaves_plan + 2 else 1) * (1 + infl)))
            chk_ae(119, dd['IRMAA-buk5'][y], (dd['IRMAA-buk5'][y-1] / (1 if y == y_spouse_leaves_plan + 2 else 1) * (1 + infl)))
            # have to convert to float from string because the dataframe uses row one for datatype, which has squirreled string
            chk_ae(120, float(dd['IRMAA-chg0'][y]), (float(dd['IRMAA-chg0'][y-1]) / (2 if y == y_spouse_leaves_plan else 1) * (1 + infl)))
            chk_ae(121, float(dd['IRMAA-chg1'][y]), (float(dd['IRMAA-chg1'][y-1]) / (2 if y == y_spouse_leaves_plan else 1) * (1 + infl)))
            chk_ae(122, float(dd['IRMAA-chg2'][y]), (float(dd['IRMAA-chg2'][y-1]) / (2 if y == y_spouse_leaves_plan else 1) * (1 + infl)))
            chk_ae(123, float(dd['IRMAA-chg3'][y]), (float(dd['IRMAA-chg3'][y-1]) / (2 if y == y_spouse_leaves_plan else 1) * (1 + infl)))
            chk_ae(124, float(dd['IRMAA-chg4'][y]), (float(dd['IRMAA-chg4'][y-1]) / (2 if y == y_spouse_leaves_plan else 1) * (1 + infl)))
            chk_ae(125, float(dd['IRMAA-chg5'][y]), (float(dd['IRMAA-chg5'][y-1]) / (2 if y == y_spouse_leaves_plan else 1) * (1 + infl)))
            chk_ae(126, dd['tax0'][y], (dd['tax0'][y-1] / (2 if y == y_spouse_leaves_plan else 1) * (1 + infl)))
            chk_ae(127, dd['tax1'][y], (dd['tax1'][y-1] / (2 if y == y_spouse_leaves_plan else 1) * (1 + infl)))
            chk_ae(128, dd['tax2'][y], (dd['tax2'][y-1] / (2 if y == y_spouse_leaves_plan else 1) * (1 + infl)))
            chk_ae(129, dd['tax3'][y], (dd['tax3'][y-1] / (2 if y == y_spouse_leaves_plan else 1) * (1 + infl)))
            chk_ae(130, dd['tax4'][y], (dd['tax4'][y-1] / (2 if y == y_spouse_leaves_plan else 1) * (1 + infl)))
            chk_ae(131, dd['tax5'][y], (dd['tax5'][y-1] / (2 if y == y_spouse_leaves_plan else 1) * (1 + infl)))
            # bracket 6 is different ratio for MFJ/Single
            r = (751.600 - 501.050) / (626.350 - 250.525)
            chk_ae(132, dd['tax6'][y], (dd['tax6'][y-1] / (r if y == y_spouse_leaves_plan else 1) * (1 + infl)))
            chk_ae(133, dd['cgt0'][y], (dd['cgt0'][y-1] / (2 if y == y_spouse_leaves_plan else 1) * (1 + infl)))
            chk_ae(134, dd['cgt15'][y], (dd['cgt15'][y-1] / (2 if y == y_spouse_leaves_plan else 1) * (1 + infl)))
            # TODO -- pre-SSA years; spouse leaves plan
            if year == 2034:
                if dd['SSA_income'][y] < dd['SSA_income'][y-1]:
                    chk_ae(135, dd['SSA_income'][y], dd['SSA_income'][y-1] * 0.77 * (1 + infl))
                else:
                    chk_ae(136, dd['SSA_income'][y], dd['SSA_income'][y-1] * (1 + infl))
            elif y == y_spouse_leaves_plan:
                chk_ae(137, 1, 1) # TODO?
            else:
                chk_ae(138, dd['SSA_income'][y], dd['SSA_income'][y-1] * (1 + infl))
            #
            chk_ae(140, dd['surplus'][y], dd['surplus'][y-1] * (1 + infl))
        
        report_sub()
    report_tot()


tst_button = widgets.Button(description='Test e-ORP', disabled=False,)
tst_button.on_click(test)

winputs = widgets.VBox([
    efname, tst_button,
    err_out,
    out_box
])

display(winputs)
```
